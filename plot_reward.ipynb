```python
import numpy as np
import re
import matplotlib.pyplot as plt

def extract_rewards(file_path):
    with open(file_path, 'r') as file:
        text = file.read()
        # The pattern matches negative floats
        pattern = r"-?\d+\.\d+"
        rewards_str = re.findall(pattern, text)
        rewards = np.array(rewards_str, dtype=float)
        return rewards

# Replace 'path_to_file.txt' with your actual file path
rewards_array = extract_rewards('/Users/john_schafer/Downloads/CE291/RNN_reward_700.txt')
rewards_array=rewards_array[1:]
timesteps = np.arange(len(rewards_array)) 
epsilon = .999 ** timesteps
```

```python

plt.figure(figsize=(14, 8))

plt.subplot(1, 2, 1)
plt.plot(timesteps, rewards_array, label='Reward', color="gold", alpha=.8)
plt.legend()
plt.title('Reward Over Episodes')
plt.xlabel('Episode Iteration')
plt.ylabel('Reward')

# Second subplot for Epsilon
plt.subplot(1, 2, 2)
plt.plot(timesteps, epsilon, label='Epsilon', color="lightblue")
plt.legend()
plt.title('Epsilon Decay')
plt.xlabel('Episode Iteration')
plt.ylabel('Epsilon')

# Show the plots
plt.tight_layout()
plt.show()
```
